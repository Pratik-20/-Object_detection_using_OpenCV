# Import Libraries

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

# Define Required Data

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model= 'frozen_inference_graph.pb'
```

# Create a Model

```python
model= cv2.dnn_DetectionModel(frozen_model, config_file)
```

# Define ClassLabels

```python
classLabels = [] ## Empty List of python
file_name = 'Labels.txt'
with open(file_name, 'rt') as fpt: 
    classLabels = fpt.read().rstrip('\n').split('\n')
    #classLables.append(fpt.read())
```

```python
print(classLabels)
```

```python
print(len(classLabels)) 
```

# Let Provide the Input size of model as mentioned in configuration file

```python
model.setInputSize(320, 320)
model.setInputScale(1.0/127.5) 
model.setInputMean((127.5, 127.5,127.5))
model.setInputSwapRB(True)
```

# Operations on Image

```python
img = cv2.imread('Input.jpeg')
#Take image as input
```

```python
plt.imshow(img)
#Show Black & White Image
```

```python
plt.imshow (cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#Show Colored Image
```

# Define ClassIndex & Confidence (Threshold Ratio/Value) 

```python
ClassIndex, confidece, bbox = model.detect(img, confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
# Object Detection
```

```python
font_scale = 3   # Size of ClassIndex Mentioned in image
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    #cv2.rectangle(frame, (x, y), (x+, yth), (255, e, e), 2)
    #cv2.putText(ng, text, (text_offset_x, text_offset y), font, fontscale-font_scale, color-(0, 0, 0), thickness-1)
    cv2.rectangle(img,boxes,(255, 0,0), 2) 
    #draw a rectangal around object
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0]+10,boxes [1]+40), font, fontScale=font_scale,color=(0, 255, 0), thickness=3)
```

# Operations on Output Image

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
#Show Output image
```

```python
cv2.imwrite('Output.jpeg',img)
#Save Output image
```
